```python
# Import libraries
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import pearsonr, iqr
from datetime import datetime
from statsmodels.robust.scale import mad as mad_c
import copy

from sklearn.linear_model import LinearRegression, HuberRegressor


import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')


# Set path for generated figures
fig_path = './figures_4.3/'

# set the scaling factor to 1
def mad(a):
    return mad_c(a, c=1)
```

```python
# Read data
mydateparser = lambda x: datetime.strptime(x, "%Y-%m-%d")

AAPL = pd.read_csv("AAPL.csv", index_col=0, parse_dates=True, date_parser=mydateparser)
IBM = pd.read_csv("IBM.csv", index_col=0, parse_dates=True, date_parser=mydateparser)
JPM = pd.read_csv("JPM.csv", index_col=0, parse_dates=True, date_parser=mydateparser)
DJI = pd.read_csv("DJI.csv", index_col=0, parse_dates=True, date_parser=mydateparser)

stocks_data = {'APPL': AAPL, 'IBM': IBM, 'JPM': JPM, 'DJI': DJI}
```

```python
for stock in stocks_data:
    stocks_data[stock]['Return'] = stocks_data[stock]['Adj Close'].pct_change()
```

## 4.2 Robust and OLS Regression

<a id='4.3.1'></a>
### 4.3.1 OLS regression

Each stock's 1-day returns are regressed against the 1-day returns of DJI using Ordinary Least Squares (OLS) regression. The regression can be defined as

\begin{equation}
    \mathbf{r}_{\text{stock}} = \mathbf{X}\mathbf{b} + \mathbf{e},
\end{equation}      

or

\begin{equation}
    \mathbf{r}_{\text{stock}} = 
    \begin{bmatrix}
    \mathbf{1} & \mathbf{r}_{\text{DJI}} \\
    \end{bmatrix}
    \begin{bmatrix}
    \alpha \\
    \beta \\
    \end{bmatrix}
    + \mathbf{e},
\end{equation}  

where $\mathbf{r}_{\text{stock}}$ is the vector representing the daily returns of the considered stock, $\mathbf{r}_{\text{DJI}}$ is the vector representing the daily returns of DJI, $\alpha$ and $\beta$ the regression coefficients, and $\mathbf{e}$ is the error vector.

The OLS method aims to minimize the squared error, given by $||\mathbf{e}||^2 = ||\mathbf{\mathbf{r}_{\text{stock}}} - \mathbf{X}\mathbf{b}||^2$. The optimal coefficients, $\hat{\mathbf{b}}$, that minimize this loss function is then given by $\hat{\mathbf{b}} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{r}_{\text{stock}}$. The regression coefficients obtained with OLS regression are given in the table below, and the predictions of this regressor are given in the figure below together with the true values of the returns. 

```python
returns = pd.DataFrame()
for stock in stocks_data:
    returns[stock] = stocks_data[stock]['Return']
returns.dropna(inplace=True)
```

```python
# OLS regression
fig, axes = plt.subplots(3,1, figsize=(9,9.5))
fig.tight_layout()
fig.subplots_adjust(hspace=0.3)

OLS_coefs = pd.DataFrame(columns=returns.columns, index=['alpha', 'beta'])
r_DJI = returns['DJI'].values.reshape(-1,1)

for i, stock in enumerate(returns.columns):
    r_stock = returns[stock].values.reshape(-1,1)
    reg = LinearRegression().fit(r_DJI, r_stock)
    
    OLS_coefs[stock]['alpha'] = float(reg.intercept_)
    OLS_coefs[stock]['beta']  = float(reg.coef_)
    
    
    if stock != 'DJI':
        ax = axes[i]
        OLS_pred = reg.predict(r_DJI)
        
        ax.plot(returns[stock], label='true')
        ax.plot(returns[stock].index, OLS_pred, label='predicted')
        ax.set_title("OLS Regression for "+stock, fontsize=20)
        ax.set_ylabel('Returns')
        ax.tick_params(labelsize=11)
        ax.legend(fontsize=12)
    
    
fig.savefig(fig_path+'OLS_Regression.eps', dpi=200, bbox_inches='tight')
```

### 4.3.2 Robust regression (Huber regression)

The Huber regression is similar to the OLS regression with a slight change in the loss function it aims to minimize. Its loss function $J$ is given by

\begin{equation}
    J(y, \hat{y}) = 
    \begin{cases}
        (y - \hat{y})^2  
        \hspace{10pt}\dots\hspace{10pt} 
        \frac{|y - \hat{y}|}{\sigma} \leq \epsilon \\
        |y - \hat{y}|    
        \hspace{17pt}\dots\hspace{10pt} 
        \frac{|y - \hat{y}|}{\sigma} > \epsilon   \\
    \end{cases}.
\end{equation}

The Huber Regressor optimizes the squared loss for the samples where $\frac{|y - \hat{y}|}{\sigma} \leq \epsilon$ and the absolute loss for the samples where $\frac{|y - \hat{y}|}{\sigma} > \epsilon$. This makes sure that the loss function is not heavily influenced by the outliers while not completely ignoring their effect.

Note that the parameter sigma makes sure that if y is scaled up
or down by a certain factor, one does not need to rescale $\epsilon$ to
achieve the same robustness. However, by changing the value of $\epsilon$, one can control the point at which an error is considered to be caused by an outlier.

The regression coefficients obtained with Huber regression are given in the table below and the predictions of this regressor are given in the figure below together with the true values of the returns.

```python
# Huber regression
fig, axes = plt.subplots(3,1, figsize=(9,9.5))
fig.tight_layout()
fig.subplots_adjust(hspace=0.3)

huber_coefs = pd.DataFrame(columns=returns.columns, index=['alpha', 'beta'])
r_DJI = returns['DJI'].values.reshape(-1,1)

for i, stock in enumerate(returns.columns):
    r_stock = returns[stock].values
    huber = HuberRegressor().fit(r_DJI, r_stock)
    
    huber_coefs[stock]['alpha'] = float(huber.intercept_)
    huber_coefs[stock]['beta']  = float(huber.coef_)
    
    if stock != 'DJI':
        ax = axes[i]
        Huber_pred = huber.predict(r_DJI)
        
        ax.plot(returns[stock], label='true')
        ax.plot(returns[stock].index, Huber_pred, label='predicted')
        ax.set_title("Huber Regression for "+stock, fontsize=20)
        ax.set_ylabel('Returns')
        ax.tick_params(labelsize=11)
        ax.legend(fontsize=12)
    

fig.savefig(fig_path+'Huber_Regression.eps', dpi=200, bbox_inches='tight')
```

### 4.3.3 Huber vs. OLS regression

```python
r_with_outliers = copy.deepcopy(returns)
dates = ['2018-05-14', '2018-09-14', '2018-12-14', '2019-01-14']

for stock in r_with_outliers.columns:
    if stock == 'DJI':
        continue
    max_val = r_with_outliers[stock].max()
    for date in dates:
        r_with_outliers[stock][date] = 1.2 * max_val
```

```python
# OLS regression
fig, axes = plt.subplots(3,1, figsize=(9,9.5))
fig.tight_layout()
fig.subplots_adjust(hspace=0.3)

returns = r_with_outliers # change returns to the version with outliers
OLS_coefs_outliers = pd.DataFrame(columns=returns.columns, index=['alpha', 'beta'])
r_DJI = returns['DJI'].values.reshape(-1,1)

for i, stock in enumerate(returns.columns):
    r_stock = returns[stock].values.reshape(-1,1)
    reg = LinearRegression().fit(r_DJI, r_stock)
    
    OLS_coefs_outliers[stock]['alpha'] = float(reg.intercept_)
    OLS_coefs_outliers[stock]['beta']  = float(reg.coef_)
    
    
    if stock != 'DJI':
        ax = axes[i]
        OLS_pred_outliers = reg.predict(r_DJI)
        
        ax.plot(returns[stock], label='true')
        ax.plot(returns[stock].index, OLS_pred_outliers, label='predicted')
        ax.set_title("OLS Regression for "+stock, fontsize=20)
        ax.set_ylabel('Returns')
        ax.tick_params(labelsize=11)
        ax.legend(fontsize=12)
    
    
# fig.savefig(fig_path+'OLS_Regression.eps', dpi=200, bbox_inches='tight')
```

```python
OLS_coefs_outliers
```

```python
# Huber regression
fig, axes = plt.subplots(3,1, figsize=(9,9.5))
fig.tight_layout()
fig.subplots_adjust(hspace=0.3)

returns = r_with_outliers # change returns to the version with outliers
huber_coefs_outliers = pd.DataFrame(columns=returns.columns, index=['alpha', 'beta'])
r_DJI = returns['DJI'].values.reshape(-1,1)

for i, stock in enumerate(returns.columns):
    r_stock = returns[stock].values
    huber = HuberRegressor().fit(r_DJI, r_stock)
    
    huber_coefs_outliers[stock]['alpha'] = float(huber.intercept_)
    huber_coefs_outliers[stock]['beta']  = float(huber.coef_)
    
    if stock != 'DJI':
        ax = axes[i]
        Huber_pred_outliers = huber.predict(r_DJI)
        
        ax.plot(returns[stock], label='true')
        ax.plot(returns[stock].index, Huber_pred_outliers, label='predicted')
        ax.set_title("Huber Regression for "+stock, fontsize=20)
        ax.set_ylabel('Returns')
        ax.tick_params(labelsize=11)
        ax.legend(fontsize=12)
    

# fig.savefig(fig_path+'Huber_Regression.eps', dpi=200, bbox_inches='tight')
```

```python
huber_coefs_outliers
```

The coefficients obtained with different regression methods have been quite similar and so was the predictions of both regressors.
```python
# # OLS and Huber together with outliers
fig, axes = plt.subplots(3,1, figsize=(10,9))
fig.tight_layout()
fig.subplots_adjust(hspace=0.3)


for i, stock in enumerate(returns.columns):
    if stock != 'DJI':
        ax = axes[i]
        
#         ax.plot(returns[stock], label='true')
        ax.plot(returns[stock].index, OLS_pred_outliers, label='OLS predicted')
        ax.plot(returns[stock].index, Huber_pred_outliers, label='Huber predicted')
        ax.set_title("Regression for "+stock+" with contaminated values", fontsize=20)
        ax.set_ylabel('Returns')
        ax.tick_params(labelsize=11)
        ax.legend(fontsize=12)
    

fig.savefig(fig_path+'Both_Regressions_Together.eps', dpi=200, bbox_inches='tight')
```

We can see in the figure above that the predictions of the different regressors are quite similar. Since we have only added four outliers, this is to be expected. However, when viewing the corresponding changes in the coefficients in the tables above after the outliers are added, we can clearly observe the robust behavior of the Huber regression. With just 4 outliers the coefficients of the OLS regression significantly change, while the coefficients of the Huber regression stay almost constant.

As mentioned before, because Huber regression, compared to OLS regression, diminishes the importance of very large errors in estimation, it is more robust to outliers. The analysis in this section empirically demonstrates this robustness.
