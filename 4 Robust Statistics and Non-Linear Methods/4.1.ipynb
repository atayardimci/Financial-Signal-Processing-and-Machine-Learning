# 4 Robust Statistics and Non-Linear Methods

In this section we will employ different robust statistical techniques on the following financial data: 3 stocks (AAPL, IBM and JPM) and 1 index (DJI); with open, high, low, close, adj. close prices; with dates ranging from 16/03/2018 to 11/03/2019.

## 4.1 Data Import and Exploratory Data Analysis

We will first conduct certain exploratory data analysis.

```python
# Import libraries
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import pearsonr, iqr
from datetime import datetime
from statsmodels.robust.scale import mad as mad_c
import copy



import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')


# Set path for generated figures
fig_path = './figures_4.1/'

# set the scaling factor to 1
def mad(a):
    return mad_c(a, c=1)
```

```python
# Read data
mydateparser = lambda x: datetime.strptime(x, "%Y-%m-%d")

AAPL = pd.read_csv("AAPL.csv", index_col=0, parse_dates=True, date_parser=mydateparser)
IBM = pd.read_csv("IBM.csv", index_col=0, parse_dates=True, date_parser=mydateparser)
JPM = pd.read_csv("JPM.csv", index_col=0, parse_dates=True, date_parser=mydateparser)
DJI = pd.read_csv("DJI.csv", index_col=0, parse_dates=True, date_parser=mydateparser)

stocks_data = {'APPL': AAPL, 'IBM': IBM, 'JPM': JPM, 'DJI': DJI}
```

<a id='4.1.1'></a>
### 4.1.1 Descriptive statistics of stocks

For each stock and for each column, we generate the key descriptive statistics that summarize the distribution of the dataset, which are mean, median, standard deviation, interquartile range, median absolute deviation, skewness (measure of asymmetry) and kurtosis (measure of the "tailedness"), given in tables below.

We also compute the 1-day returns, using the adj. close column for each stock, and add them to their corresponding dataframe as a new column.

In general, means and medians are very close to each other within columns, but standard deviation and median absolute deviations can largely differ. We observe that in general median absolute deviations are smaller than standard deviations. We also note that the median absolute deviation is generally close to the half of interquartile range.

```python
stocks_stats = {}

for stock in stocks_data:
    df = stocks_data[stock]
    stats = pd.DataFrame(columns=df.columns)

    stats.loc['mean'] = df.mean()
    stats.loc['median'] = df.median()
    stats.loc['stddev'] = df.std()
    stats.loc['MAD'] = mad(df)
    stats.loc['IQR'] = iqr(df, axis=0)
    stats.loc['skewness'] = df.skew()
    stats.loc['kurtosis'] = df.kurtosis()

    stocks_stats[stock] = copy.deepcopy(stats)
```

```python
for stock in stocks_data:
    stocks_data[stock]['Returns'] = stocks_data[stock]['Adj Close'].pct_change()
```

### 4.1.2 Pdf of adjusted close and returns

In figures below, we plot the empirical probability density functions of adjusted close prices and 1-day returns, using histograms. As discussed in section 1.1.3, the returns tend to follow a more Gaussian distribution then the prices, and this can be clearly seen in the distributions below. 

When a distribution of a signal is not Gaussian, it may not be appropriate to represent that signal with its mean and standard deviation; it may be more appropriate to use statistics such as the median and median absolute deviation in terms of robustness.

```python
fig, axes = plt.subplots(2,4, figsize=(18,6))
fig.tight_layout()
fig.subplots_adjust(hspace=0.3, wspace=0.3)


for i, stock in enumerate(stocks_data):
    adj_close = stocks_data[stock]['Adj Close']
    returns = stocks_data[stock]['Returns']
    
    # Plot adjusted close
    axes[0,i].set_title(stock+' Adj Close', fontsize=15)
    counts_and_vals = axes[0,i].hist(adj_close, bins=30, color='darkgoldenrod');
    
    axes[0,i].tick_params(axis='y', colors='darkgoldenrod', labelsize=12)
    axes[0,i].tick_params(axis='x', labelsize=12)


    x_vals = (counts_and_vals[1][1:] + counts_and_vals[1][0:-1]) / 2
    y_vals = counts_and_vals[0]/counts_and_vals[0].sum()
    
    twinax = axes[0,i].twinx()
    twinax.plot(x_vals, y_vals, color='midnightblue')
    twinax.tick_params(axis='y', colors='midnightblue', labelsize=12)
    twinax.grid(False)


    # Plot returns
    axes[1,i].set_title(stock+' Returns', fontsize=15)
    counts_and_vals = axes[1,i].hist(returns, bins=30, color='darkgoldenrod');

    axes[1,i].tick_params(axis='y', colors='darkgoldenrod', labelsize=12)
    axes[1,i].tick_params(axis='x', labelsize=12)
    vals = axes[1,i].get_xticks()
    axes[1,i].set_xticklabels(['{:,.0%}'.format(x) for x in vals]);

    
    x_vals = (counts_and_vals[1][1:] + counts_and_vals[1][0:-1]) / 2
    y_vals = counts_and_vals[0]/counts_and_vals[0].sum()
    
    twinax = axes[1,i].twinx()
    twinax.plot(x_vals, y_vals, color='midnightblue')
    twinax.tick_params(axis='y', colors='midnightblue', labelsize=12)
    twinax.grid(False)
    

fig.savefig(fig_path+'PDFs.eps', dpi=200, bbox_inches='tight')
```

### 4.1.3 Rolling mean and median statistics

In a figure below, for each stock we plot the adj. close, the associated **rolling mean** (using a 5-day window), and the $\pm 1.5 \times $ **standard deviations** relative to the rolling mean (first method); then in a following figure, we repeat these steps using the **rolling median** and $\pm 1.5 \times $ **median absolute deviation** relative to the rolling median (second method).

The relative rolling mean and median plots in red are quite similar to each other. However, we observe in the figures that the second method is tighter than the first in terms of rolling deviations. This is further supported by the table below as there are more points outside of the given rolling area in the median method, suggesting that the mean estimators are more susceptible to changes than the median estimators. 

```python
outliers = pd.DataFrame(index=stocks_data.keys(), columns=['mean method', 'median method'])
```

```python
# Rolling mean and std
window = 5

fig, axes = plt.subplots(2,2, figsize=(14,8))
fig.tight_layout()
fig.subplots_adjust(hspace=0.4, wspace=0.15)


for i, stock in enumerate(stocks_data):
    adj_close = stocks_data[stock]['Adj Close']

    rolling_mean = adj_close.rolling(window).mean()
    rolling_std  = adj_close.rolling(window).std()

    std_upper = rolling_mean + 1.5*rolling_std
    std_below = rolling_mean - 1.5*rolling_std

    if i == 0:
        ax = axes[0,0]
    elif i == 1:
        ax = axes[0,1]
    elif i == 2:
        ax = axes[1,0]
    elif i == 3:
        ax = axes[1,1]
    
    adj_close.plot(ax=ax, color='midnightblue', label='Adj Close', fontsize=12)
    rolling_mean.plot(ax=ax, color='firebrick', label='MA({})'.format(window))
    ax.fill_between(std_upper.index, std_upper, std_below, alpha=0.35, color='darkgoldenrod', label='\u00B1 1.5 std')
    
    ax.legend()
    ax.set_title(stock, fontsize=16)
    ax.set_ylabel('Price')
    ax.set_xlabel('')
    
    outliers['mean method'][stock] = (adj_close > std_upper).sum() + (adj_close < std_below).sum()
    

fig.savefig(fig_path+'Rolling_mean.eps', dpi=200, bbox_inches='tight')
```

```python
# Rolling median and mad
window = 5

fig, axes = plt.subplots(2,2, figsize=(14,8))
fig.tight_layout()
fig.subplots_adjust(hspace=0.4, wspace=0.15)


for i, stock in enumerate(stocks_data):
    adj_close = stocks_data[stock]['Adj Close']

    rolling_median = adj_close.rolling(window).median()
    rolling_mad = adj_close.rolling(window).apply(mad, raw=True)

    mad_upper = rolling_median + 1.5*rolling_mad
    mad_below = rolling_median - 1.5*rolling_mad

    if i == 0:
        ax = axes[0,0]
    elif i == 1:
        ax = axes[0,1]
    elif i == 2:
        ax = axes[1,0]
    elif i == 3:
        ax = axes[1,1]
    
    adj_close.plot(ax=ax, color='midnightblue', label='Adj Close', fontsize=12)
    rolling_median.plot(ax=ax, color='firebrick', label='MM({})'.format(window))
    ax.fill_between(mad_upper.index, mad_upper, mad_below, alpha=0.35, color='darkgoldenrod', label='\u00B1 1.5 std')
    
    ax.legend()
    ax.set_title(stock, fontsize=16)
    ax.set_ylabel('Price')
    ax.set_xlabel('');
    
    outliers['median method'][stock] = (adj_close > mad_upper).sum() + (adj_close < mad_below).sum() 
    
    
    
fig.savefig(fig_path+'Rolling_median.eps', dpi=200, bbox_inches='tight')
```

### 4.1.4 Robustness to outliers

We introduce four outlier points at certain dates with a value equal to 1.2 $\times$ the maximum value. It is clear that the rolling mean plot is largely affected by the outliers while the rolling median doesn't seem to have changed at all (plots in red). Furthermore, it is clear that the standard deviations are largely affected by the outliers as well, while the median absolute deviations stay robust against them. 

The points above suggest the robustness of median estimators against contaminated data and the susceptibility of mean estimators towards outliers.

```python
dates = ['2018-05-14', '2018-09-14', '2018-12-14', '2019-01-14']

adj_close_df = {}

for stock in stocks_data:
    adj_close_df[stock] = stocks_data[stock]['Adj Close']

adj_close_df = pd.DataFrame(adj_close_df)

for stock in adj_close_df.columns:
    max_val = adj_close_df[stock].max()
    for date in dates:
        adj_close_df[stock][date] = 1.2 * max_val
```

```python
outliers = pd.DataFrame(index=adj_close_df.columns, columns=['mean method', 'median method'])
```

```python
# Rolling mean and std with added outliers
window = 5

fig, axes = plt.subplots(2,2, figsize=(14,8))
fig.tight_layout()
fig.subplots_adjust(hspace=0.4, wspace=0.15)


for i, stock in enumerate(adj_close_df.columns):
    adj_close = adj_close_df[stock]

    rolling_mean = adj_close.rolling(window).mean()
    rolling_std  = adj_close.rolling(window).std()

    std_upper = rolling_mean + 1.5*rolling_std
    std_below = rolling_mean - 1.5*rolling_std

    if i == 0:
        ax = axes[0,0]
    elif i == 1:
        ax = axes[0,1]
    elif i == 2:
        ax = axes[1,0]
    elif i == 3:
        ax = axes[1,1]
    
    adj_close.plot(ax=ax, color='midnightblue', label='Adj Close', fontsize=12)
    rolling_mean.plot(ax=ax, color='firebrick', label='MA({})'.format(window))
    ax.fill_between(std_upper.index, std_upper, std_below, alpha=0.35, color='darkgoldenrod', label='\u00B1 1.5 std')
    
    ax.legend()
    ax.set_title(stock, fontsize=16)
    ax.set_ylabel('Price')
    ax.set_xlabel('')
    
    outliers['mean method'][stock] = (adj_close > std_upper).sum() + (adj_close < std_below).sum()


fig.savefig(fig_path+'Rolling_mean_with_outliers.eps', dpi=200, bbox_inches='tight')
```

```python
# Rolling median and mad with added outliers
window = 5

fig, axes = plt.subplots(2,2, figsize=(14,8))
fig.tight_layout()
fig.subplots_adjust(hspace=0.4, wspace=0.15)


for i, stock in enumerate(adj_close_df.columns):
    adj_close = adj_close_df[stock]

    rolling_median = adj_close.rolling(window).median()
    rolling_mad = adj_close.rolling(window).apply(mad, raw=True)

    mad_upper = rolling_median + 1.5*rolling_mad
    mad_below = rolling_median - 1.5*rolling_mad

    if i == 0:
        ax = axes[0,0]
    elif i == 1:
        ax = axes[0,1]
    elif i == 2:
        ax = axes[1,0]
    elif i == 3:
        ax = axes[1,1]
    
    adj_close.plot(ax=ax, color='midnightblue', label='Adj Close', fontsize=12)
    rolling_median.plot(ax=ax, color='firebrick', label='MM({})'.format(window))
    ax.fill_between(mad_upper.index, mad_upper, mad_below, alpha=0.35, color='darkgoldenrod', label='\u00B1 1.5 std')
    
    ax.legend()
    ax.set_title(stock, fontsize=16)
    ax.set_ylabel('Price')
    ax.set_xlabel('');
    
    outliers['median method'][stock] = (adj_close > mad_upper).sum() + (adj_close < mad_below).sum() 
    

fig.savefig(fig_path+'Rolling_median_with_outliers.eps', dpi=200, bbox_inches='tight')
```

### 4.1.5 Box plots of prices

Another useful tool in visualising the distribution, dispersion and key statistics of the data is the box plot. We have the box plot of close prices of each asset in the figure below.

The blue line denotes the median, the rectangular box represents the interquartile range (middle 50\% of the data), and the whiskers extend to the total range of the data, excluding the outliers which are points outside 1.5 $\times$ IQR $\pm$ 0.75 and 0.25 quartile.

In the figure below, we can see that the plots are skewed with asymmetric quartiles. This further supports the argument that it would be inappropriate to model the distribution of the prices with a Gaussian model.

```python
fig, axes = plt.subplots(1,4, figsize=(12,4))
fig.tight_layout()
fig.subplots_adjust(hspace=0.2, wspace=0.4)


for i, stock in enumerate(stocks_data):
    adj_close = stocks_data[stock]['Adj Close']

    ax = axes[i]
    
    ax.boxplot(adj_close)
    
    ax.set_title(stock, fontsize=14)
    ax.set_ylabel('Price')
    ax.set_xticks([])
    

fig.savefig(fig_path+'Box_plots.eps', dpi=200, bbox_inches='tight')
```
