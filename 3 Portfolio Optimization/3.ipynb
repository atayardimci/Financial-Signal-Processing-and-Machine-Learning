# 3 Portfolio Optimization

## 3.1 Adaptive Minimum-Variance Portfolio Optimization

Consider a portfolio of $N$ assets where the price of an asset $i$ at time $t$ is denoted by $p_i[t]$. The return of each asset at time $t$, $r_i[t]$, is then given by

\begin{equation}
    r_i[t] = \frac{p_i[t] - p_i[t-1]}{p_i[t-1]}.
\end{equation}

It is convenient to collect the returns together into a vector, $\mathbf{r}[t] = [r_1[t], \dots, r_N[t]]^T$, where each asset return is characterised by its mean and variance, i.e., $\mu_i = E\{r_i[t]\}$ and $\sigma_i^2 = E\{(r_i[t] - \mu_i)^2\}$. The correlation between the returns of two assets $i$ and $j$ is given by the correlation coefficient $\rho_{ij} \in [-1, 1]$, while the covariance matrix of all assets is given by

\begin{equation}
    \boldsymbol{\Sigma} = E\{ (\mathbf{r}[t] - \boldsymbol{\mu}) (\mathbf{r}[t] - \boldsymbol{\mu})^T \} \in \mathbb{R}^{N \times N},
\end{equation}

where $\boldsymbol{\mu} = E\{\mathbf{r}[t]\} \in \mathbb{R}^N$. The covariance matrix summarizes the "risk" structure of the system, where the individual covariances between assets $i$ and $j$ are given by 
$\boldsymbol{\Sigma}_{ij} = \sigma_{ij} = \rho_{ij}\sigma_i\sigma_j$.

The holdings of a portfolio can be represented by a set of weights, $\mathbf{w} = [w_1,\dots,w_N]^T$, that are applied to assets. In other words, the weight $w_i$ represents the percentage of our capital that is allocated to asset $i$. The return of the portfolio at time $t$ is therefore given by $r_P[t] = \mathbf{w}^T\mathbf{r}[t]$, with its expected return and variance respectively given as

\begin{equation}
    \mu_P = E\{r_P[t]\} = \mathbf{w}^T\boldsymbol{\mu},
\end{equation}

<a id='eq:46'></a>
\begin{equation} \label{eq:43}
    \sigma_P^2 = E \{(r_P[t] - \mu_P)^2\} = 
    \mathbf{w}^t \boldsymbol{\Sigma} \mathbf{w}.
\end{equation}
Finding the set of weights $\mathbf{w}$ for a portfolio represents a large portion of the task of asset managers. One of the simplest criteria used in portfolio management is to minimize the risk (measured by variance) of the portfolio, given by Equation [(43)](#eq:43). 

<a id='3.1.1'></a>
### 3.1.1 Optimal weights of the minimum-variance portfolio

The minimum variance portfolio is derived based on finding the set of optimal weights, $\mathbf{w}$, which minimizes the variance, $\sigma_P^2$, subject to the portfolio weights summing up to unity, i.e., $\mathbf{w}^T \mathbf{1} = 1$. This can be expressed as an optimization problem in the following way:

\begin{align*}
    \text{minimize} \hspace{20pt} \frac{1}{2} 
    &\mathbf{w}^T \boldsymbol{\Sigma} \mathbf{w}\\
    \text{subject to} \hspace{22pt}
    &\mathbf{w}^T \mathbf{1} = 1,
\end{align*}

or, equivalently, as the Lagrange optimization

\begin{equation} \label{eq:44}
    \min_{\mathbf{w}, \lambda} 
    J(\mathbf{w}, \lambda) = 
    \frac{1}{2} \mathbf{w}^T \boldsymbol{\Sigma} \mathbf{w}
    - \lambda(\mathbf{w}^T\mathbf{1} - 1).
\end{equation}

Then from the Lagrangian, we obtain the following optimality conditions:

<a id='eq:46'></a>
\begin{align}
    \frac{\partial J}{\partial \mathbf{w}} = \mathbf{0}
    \hspace{15pt} &\Rightarrow \hspace{15pt}
    \boldsymbol{\Sigma} \mathbf{w} - \lambda\mathbf{1} = \mathbf{0} \label{eq:45} \\
    \frac{\partial J}{\partial \lambda} = 0 
    \hspace{15pt} &\Rightarrow \hspace{15pt}
    \mathbf{w}^T\mathbf{1} = 1
\end{align}

We can rewrite the optimality conditions in the following form:

\begin{equation}
    \begin{bmatrix}
    \boldsymbol{\Sigma} & -\mathbf{1} \\
    -\mathbf{1}^T & 0 \\
    \end{bmatrix}
    \begin{bmatrix}
    \mathbf{w} \\
    \lambda \\
    \end{bmatrix} 
    =
    \begin{bmatrix}
    \mathbf{0} \\ 
    - 1 \\
    \end{bmatrix}
\end{equation}

Then the values for $\mathbf{w}$ and $\lambda$ are given by

\begin{equation}
    \begin{bmatrix}
    \mathbf{w} \\
    \lambda \\
    \end{bmatrix} 
    =
    \begin{bmatrix}
    \boldsymbol{\Sigma} & -\mathbf{1} \\
    -\mathbf{1}^T & 0 \\
    \end{bmatrix}^{-1}
    \begin{bmatrix}
    \mathbf{0} \\ 
    - 1 \\
    \end{bmatrix}
\end{equation}

We can also compute $\mathbf{w}$ the following way: we have 

<a id='eq:49'></a>
\begin{equation} \label{eq:49}
    \mathbf{w} = \lambda \boldsymbol{\Sigma}^{-1} \mathbf{1}.
\end{equation}

Plugging this value of $\mathbf{w}$ into Equation [(46)](#eq:46), we have

\begin{equation}
    \lambda \mathbf{1}^T \boldsymbol{\Sigma}^{-1} \mathbf{1} = 1
    \hspace{15pt} \Rightarrow \hspace{15pt}
    \lambda = \frac{1}{\mathbf{1}^T \boldsymbol{\Sigma}^{-1} \mathbf{1}}
\end{equation}

Plugging this value of $\lambda$ back into Equation [(49)](#eq:49), we compute the optimal set of weights as

\begin{equation}
    \mathbf{w}_{\text{opt}} = \frac{\boldsymbol{\Sigma}^{-1} \mathbf{1}}{\mathbf{1}^T \boldsymbol{\Sigma}^{-1} \mathbf{1}}.
\end{equation}

The variance of the portfolio constructed with this set of weights is then given by

\begin{align}
    \sigma_P^2 &= 
    \mathbf{w}_{\text{opt}}^T \boldsymbol{\Sigma} \mathbf{w}_{\text{opt}} \nonumber \\
    &= \frac{\mathbf{1}^T (\boldsymbol{\Sigma}^{-1})^T 
    \boldsymbol{\Sigma} 
    \boldsymbol{\Sigma}^{-1} \mathbf{1}}
    {(\mathbf{1}^T \boldsymbol{\Sigma}^{-1} \mathbf{1})^2} \nonumber \\
    &= \frac{\mathbf{1}^T (\boldsymbol{\Sigma}^{-1})^T \mathbf{1}}
    {(\mathbf{1}^T \boldsymbol{\Sigma}^{-1} \mathbf{1})^2} \nonumber \\
    &= \frac{1}
    {\mathbf{1}^T \boldsymbol{\Sigma}^{-1} \mathbf{1}},
\end{align}

noting that $\boldsymbol{\Sigma}$ is a positive semi-definite matrix. Also note that $\sigma_P^2 = \lambda$.

Finally, we note that the problem above doesn't put a constraint on the return target of the portfolio but only aims to find the portfolio with the minimum variance no matter the return. The problem also doesn't put a constraint on the signs of the weights, that is, the weights can be negative, which means short selling is allowed in the universe of this problem.

### 3.1.2 Minimum-variance portfolio vs. equally weighted portfolio

```python
# Import libraries
import numpy as np
import pandas as pd
from scipy import stats
from datetime import datetime

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

```

```python
# Read data
mydateparser = lambda x: datetime.strptime(x, "%Y-%m-%d")
df = pd.read_csv("fsp_case_31_BSD.csv", index_col=0, header=[0,1], parse_dates=True, date_parser=mydateparser)
```

We use the daily return data of 10 stocks for 2017 and 2018 to construct a minimum-variance (MV) portfolio. We first split this data into training and testing data, that is, we use the data for 2017 for training and 2018 for testing. Then we compute the covariance matrix for the returns of 2017 and use it to compute the optimal set of weights, $\mathbf{w_\text{opt}}$ as in the previous section.

We apply $\mathbf{w_\text{opt}}$ both to training data and to test data to evaluate the performance of the MV portfolio and plot the resulting returns and cumulative returns in figures below and summarize certain performance statistics in multiple tables. We then do the same analysis for an equally-weighted portfolio, that is $w_i = 1/10$ for $i = 1,\dots,10$.

```python
returns = df['ret'].dropna(1).iloc[:,-10:]

# split the data into train and test
returns_train = returns.iloc[:len(returns)//2] # 2017 data
returns_test  = returns.iloc[len(returns)//2:] # 2018 data
```

```python
# Compute optimal weights for the minimum variance portfolio
Sigma = returns_train.cov()
ones = np.ones((returns_train.shape[1],1))

Lambda = float(1 / ones.T.dot(np.linalg.inv(Sigma)).dot(ones))
w_opt = Lambda * np.linalg.inv(Sigma).dot(ones)
```

```python
print('Theoretical variance of the MV portfolio: {:.6f}'.format(Lambda))
```

```python
# Evaluate the performance of the MV portfolio on the train data
mv_portfolio = (returns_train * w_opt.T).sum(axis=1)

fig, axes = plt.subplots(2,1, figsize=(7,6))
fig.tight_layout()
fig.subplots_adjust(hspace=0.3)

yticks = np.arange(-0.02, 0.03, 0.01)
mv_portfolio.plot(ax=axes[0], fontsize=12, yticks=yticks); axes[0].set_title("Minimum-Variance Portfolio Daily Returns (Train)", fontsize=16)
mv_portfolio.cumsum().plot(ax=axes[1], fontsize=12); axes[1].set_title("Minimum-Variance Portfolio Cumulative Return (Train)", fontsize=16)

mv_portfolio_stats = pd.Series(dtype=float)
mv_portfolio_stats['mean return'] = mv_portfolio.mean()
mv_portfolio_stats['cumulative return'] = mv_portfolio.cumsum()[-1]
mv_portfolio_stats['variance of returns'] = mv_portfolio.std()**2
mv_portfolio_stats['sharpe ratio'] = mv_portfolio.mean() / mv_portfolio.std()
print(mv_portfolio_stats)

fig.savefig('MV_Portfolio_Train_Performance.eps', dpi=200, bbox_inches='tight')
```

```python
# Evaluate the performance of the MV portfolio on the test data
mv_portfolio = (returns_test * w_opt.T).sum(axis=1)

fig, axes = plt.subplots(2,1, figsize=(7,6))
fig.tight_layout()
fig.subplots_adjust(hspace=0.3)

mv_portfolio.plot(ax=axes[0], fontsize=12); axes[0].set_title("Minimum-Variance Portfolio Daily Returns (Test)", fontsize=16)
mv_portfolio.cumsum().plot(ax=axes[1], fontsize=12); axes[1].set_title("Minimum-Variance Portfolio Cumulative Return (Test)", fontsize=16)

mv_portfolio_stats = pd.Series(dtype=float)
mv_portfolio_stats['mean return'] = mv_portfolio.mean()
mv_portfolio_stats['cumulative return'] = mv_portfolio.cumsum()[-1]
mv_portfolio_stats['variance of returns'] = mv_portfolio.std()**2
mv_portfolio_stats['sharpe ratio'] = mv_portfolio.mean() / mv_portfolio.std()
print(mv_portfolio_stats)

fig.savefig('MV_Portfolio_Test_Performance.eps', dpi=200, bbox_inches='tight')
```

```python
# Evaluate the performance of the equally-weighted portfolio on the train data
mv_portfolio = returns_train.mean(1)

fig, axes = plt.subplots(2,1, figsize=(7,6))
fig.tight_layout()
fig.subplots_adjust(hspace=0.3)

mv_portfolio.plot(ax=axes[0], fontsize=12); axes[0].set_title("Equally-Weighted Portfolio Daily Returns (Train)", fontsize=16)
mv_portfolio.cumsum().plot(ax=axes[1], fontsize=12); axes[1].set_title("Equally-Weighted Portfolio Cumulative Return (Train)", fontsize=16)

mv_portfolio_stats = pd.Series(dtype=float)
mv_portfolio_stats['mean return'] = mv_portfolio.mean()
mv_portfolio_stats['cumulative return'] = mv_portfolio.cumsum()[-1]
mv_portfolio_stats['variance of returns'] = mv_portfolio.std()**2
mv_portfolio_stats['sharpe ratio'] = mv_portfolio.mean() / mv_portfolio.std()
print(mv_portfolio_stats)

fig.savefig('EW_Portfolio_Train_Performance.eps', dpi=200, bbox_inches='tight')
```

```python
# Evaluate the performance of the equally-weighted portfolio on the test data
mv_portfolio = returns_test.mean(1)

fig, axes = plt.subplots(2,1, figsize=(7,6))
fig.tight_layout()
fig.subplots_adjust(hspace=0.3)

mv_portfolio.plot(ax=axes[0], fontsize=12); axes[0].set_title("Equally-Weighted Portfolio Daily Returns (Test)", fontsize=16)
mv_portfolio.cumsum().plot(ax=axes[1], fontsize=12); axes[1].set_title("Equally-Weighted Portfolio Cumulative Return (Test)", fontsize=16)

mv_portfolio_stats = pd.Series(dtype=float)
mv_portfolio_stats['mean return'] = mv_portfolio.mean()
mv_portfolio_stats['cumulative return'] = mv_portfolio.cumsum()[-1]
mv_portfolio_stats['variance of returns'] = mv_portfolio.std()**2
mv_portfolio_stats['sharpe ratio'] = mv_portfolio.mean() / mv_portfolio.std()
print(mv_portfolio_stats)

fig.savefig('EW_Portfolio_Test_Performance.eps', dpi=200, bbox_inches='tight')
```

We observe that the returns of the MV portfolio during the training period exhibit the smallest variance as expected with a value of $2.9 \times 10^{-5}$, which is also equal to the theoretical variance value that was computed using the equation from the previous section. However, the MV portfolio performs worse than the EW portfolio in terms of volatility when applied to test data. This indicates the unsteady nature of the financial markets and suggests the use of a more sophisticated model.

We also note the upward trend in cumulative returns during the training period and the downward trend during the test period (for both constructed portfolios), which indicates the mean-reverting nature of the stocks.

### 3.1.3 Adaptive minimum-variance portfolio

In this section we implement an adaptive time-varying minimum-variance portfolio, using a rolling window of length $M$ days. In the figure below, we plot the performance of adaptive portfolios constructed with varying window lengths. 

We observe a clear downward trend in cumulative returns as the window length increases. This suggests the use of more recent return data to compute the portfolio weights performs better in terms of the cumulative return the portfolio yields. However, we also observe an improvement in performance in terms of variance as the window length increases, although we note that the variance doesn't decrease much beyond a 2 months (44 days) window.

We also plot the daily returns, cumulative returns, and the value of the rolling weights for window lengths of 252, 66, and 22 days for comparison in other figures, and summarize their performance statistics in tables alongside the statistics from the portfolios (mean-variance and equally weighted portfolios) from section [3.1.2](#3.1.2).

```python
def compute_w_opt(Sigma):
    ones = np.ones((Sigma.shape[0],1))
    Lambda = float(1 / ones.T.dot(np.linalg.inv(Sigma)).dot(ones))
    w_opt = Lambda * np.linalg.inv(Sigma).dot(ones)
    return w_opt.reshape(-1,)
```

```python
def compute_portfolio_and_rolling_w(returns, dates, window):
    weights_df = pd.DataFrame(columns=returns.columns) 
    adaptive_mv_portfolio = pd.Series(dtype=float)

    for t in dates:
        Sigma_t = returns[:t].iloc[-window-1:-1].cov()
        w_t = compute_w_opt(Sigma_t)
        weights_df.loc[t] = w_t
        adaptive_mv_portfolio[t] = (returns.loc[t] * w_t).sum()
    
    return (adaptive_mv_portfolio, weights_df)
```

```python
# Adaptive MV portfolio with a 252-day window
adaptive_mv_portfolio, weights_df = compute_portfolio_and_rolling_w(returns, returns_test.index, 252)

fig, axes = plt.subplots(2,2, figsize=(12,4))

axes[0,1].remove()
axes[1,1].remove()
ax_right = fig.add_subplot(2,2,(2,4))

fig.tight_layout()
fig.subplots_adjust(hspace=0.42)

adaptive_mv_portfolio.plot(ax=axes[0,0], fontsize=12); axes[0,0].set_title("Adaptive MV Portfolio Daily Returns (M=252)", fontsize=14)
adaptive_mv_portfolio.cumsum().plot(ax=axes[1,0], fontsize=12); axes[1,0].set_title("Adaptive MV Portfolio Cumulative Return (M=252)", fontsize=14)
weights_df.plot(ax=ax_right, fontsize=12); ax_right.set_title("Rolling Weights (M=252)", fontsize=14)
ax_right.legend(loc='upper center', bbox_to_anchor=(0.4, 1), ncol=3)

portfolio_stats = pd.Series(dtype=float)
portfolio_stats['mean return'] = adaptive_mv_portfolio.mean()
portfolio_stats['cumulative return'] = adaptive_mv_portfolio.cumsum()[-1]
portfolio_stats['variance of returns'] = adaptive_mv_portfolio.std()**2
portfolio_stats['sharpe ratio'] = adaptive_mv_portfolio.mean() / adaptive_mv_portfolio.std()
print(portfolio_stats)

fig.savefig('Adaptive_MV_Portfolio_252.eps', dpi=200, bbox_inches='tight')
```

```python
# Adaptive MV portfolio with a 66-day window
adaptive_mv_portfolio, weights_df = compute_portfolio_and_rolling_w(returns, returns_test.index, 66)

fig, axes = plt.subplots(2,2, figsize=(12,4))

axes[0,1].remove()
axes[1,1].remove()
ax_right = fig.add_subplot(2,2,(2,4))

fig.tight_layout()
fig.subplots_adjust(hspace=0.42)

adaptive_mv_portfolio.plot(ax=axes[0,0], fontsize=12); axes[0,0].set_title("Adaptive MV Portfolio Daily Returns (M=66)", fontsize=14)
adaptive_mv_portfolio.cumsum().plot(ax=axes[1,0], fontsize=12); axes[1,0].set_title("Adaptive MV Portfolio Cumulative Return (M=66)", fontsize=14)
weights_df.plot(ax=ax_right, fontsize=12); ax_right.set_title("Rolling Weights (M=66)", fontsize=14)
ax_right.legend(loc='upper center', bbox_to_anchor=(0.4, 1), ncol=3)

portfolio_stats = pd.Series(dtype=float)
portfolio_stats['mean return'] = adaptive_mv_portfolio.mean()
portfolio_stats['cumulative return'] = adaptive_mv_portfolio.cumsum()[-1]
portfolio_stats['variance of returns'] = adaptive_mv_portfolio.std()**2
portfolio_stats['sharpe ratio'] = adaptive_mv_portfolio.mean() / adaptive_mv_portfolio.std()
print(portfolio_stats)

fig.savefig('Adaptive_MV_Portfolio_66.eps', dpi=200, bbox_inches='tight')
```

```python
# Adaptive MV portfolio with a 22-day window
adaptive_mv_portfolio, weights_df = compute_portfolio_and_rolling_w(returns, returns_test.index, 22)

fig, axes = plt.subplots(2,2, figsize=(12,4))

axes[0,1].remove()
axes[1,1].remove()
ax_right = fig.add_subplot(2,2,(2,4))

fig.tight_layout()
fig.subplots_adjust(hspace=0.42)

adaptive_mv_portfolio.plot(ax=axes[0,0], fontsize=12); axes[0,0].set_title("Adaptive MV Portfolio Daily Returns (M=22)", fontsize=14)
adaptive_mv_portfolio.cumsum().plot(ax=axes[1,0], fontsize=12); axes[1,0].set_title("Adaptive MV Portfolio Cumulative Return (M=22)", fontsize=14)
weights_df.plot(ax=ax_right, fontsize=12); ax_right.set_title("Rolling Weights (M=22)", fontsize=14)
ax_right.legend(loc='upper center', bbox_to_anchor=(0.4, 1), ncol=3)

portfolio_stats = pd.Series(dtype=float)
portfolio_stats['mean return'] = adaptive_mv_portfolio.mean()
portfolio_stats['cumulative return'] = adaptive_mv_portfolio.cumsum()[-1]
portfolio_stats['variance of returns'] = adaptive_mv_portfolio.std()**2
portfolio_stats['sharpe ratio'] = adaptive_mv_portfolio.mean() / adaptive_mv_portfolio.std()
print(portfolio_stats)

fig.savefig('Adaptive_MV_Portfolio_22.eps', dpi=200, bbox_inches='tight')
```

```python
n_months = [i for i in range(1,13)]
cumulative_returns = []
variances = []

for i in n_months:
    adaptive_mv_portfolio, weights_df = compute_portfolio_and_rolling_w(returns, returns_test.index, 22*i)
    cumulative_returns.append(adaptive_mv_portfolio.cumsum()[-1])
    variances.append(adaptive_mv_portfolio.std()**2)
```

```python
fig, axes = plt.subplots(2,1, figsize=(7,5))
fig.tight_layout()
fig.subplots_adjust(hspace=0.5)

axes[0].bar(n_months, cumulative_returns)
axes[0].set_title('Cumulative Return', fontsize=16)
axes[0].set_xlabel('Window Lenght (months)')


axes[1].bar(n_months, variances)
axes[1].set_title('Variance', fontsize=16)
axes[1].set_xlabel('Window Lenght (months)')


fig.savefig('Window_Length_vs_Performance.eps', dpi=200, bbox_inches='tight')
```

Overall performance of adaptive portfolios is better the performances of static minimum-variance and equally weighted portfolios in section [3.2.1](#3.1.2); only the adaptive MV portfolio with a rolling window of 22 days performs worse in terms of volatility.

We also observe that even the volatility of returns decrease with increasing window size, the returns become mostly negative. The way we constructed our optimization problem was to find the portfolio with the minimum variance and we didn't take into account the actual values of returns, so a poor performance in terms of return magnitude is expected.


Finally, in this section we compute the sample covariance matrix of returns using the most recent $M$ returns of each company using the following calculation method:

\begin{equation}
    \hat{\boldsymbol{\Sigma}}[t] = \frac{1}{M} 
    \sum_{\tau=t-M}^{t-1} 
    (\mathbf{r}[\tau] - \hat{\boldsymbol{\mu}}[t]) 
    (\mathbf{r}[\tau] - \hat{\boldsymbol{\mu}}[t])^T.
\end{equation}

We can change this calculation slightly by assigning different weights to different days. For instance, we can allocate the weights in an exponentially weighted manner to put more emphasis on more recent returns. This can further enhance the model.
